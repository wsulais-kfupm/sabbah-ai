```python
import asteroid
import soundfile as sf
import librosa
import numpy

vsc_model_name = "JorisCos/ConvTasNet_Libri2Mix_sepnoisy_8k"
vsc_model = asteroid.models.conv_tasnet.ConvTasNet.from_pretrained(vsc_model_name)

def split_voices(audio_path):
    mix, mix_sr = sf.read(audio_path)
    mix = librosa.resample(mix, orig_sr=mix_sr, target_sr=8000)
    mix = mix.transpose()
    mix = mix.reshape(1, *mix.shape)
    out_wavs = vsc_model.separate(mix.astype(numpy.float32))
    out_wavs = librosa.resample(out_wavs, orig_sr=8000, target_sr=16_000)
    return out_wavs
    
```

```python
#model_HSC="/Users/malik/Library/CloudStorage/GoogleDrive-abdulmalikmeshari@gmail.com/My Drive/HSC-AT"
model_HSC="/Users/wael/Documents/code/python/torch-nix-venv/HSC-AT"
model_audio = "openai/whisper-medium"
model_audio = "/Users/wael/checkpoint-3000"
audio_ex="/Users/malik/Desktop/wael.mp3"
```

```python
pipe_HSC=pipeline('text-classification', model=model_HSC)
pipe_audio=pipeline(task="automatic-speech-recognition", model=model_audio,)
```

```python
def predict_text(text):
    result=pipe_HSC(text)[0]
    score=result["score"] * 100
    if result['label'] == 'LABEL_0':
        return f'The text is non-hateful with a {score:.01f}% certainty'
    return f'The text is hateful with a {score:.01f}% certainty'



def predict_audio(audio):
    text=pipe_audio(audio, generate_kwargs={ "task": "transcribe", "language": "<|ar|>" })['text']
    result=pipe_HSC(text)[0]
    score=result["score"] * 100
    if result['label'] == 'LABEL_0':
        return text, f'The text transcribed non-hateful with a {score:.01f}% certainty'
    return text, f'The text transcribed is hateful with a {score:.01f}% certainty'

def describe_result(result):
    score = result["score"] * 100
    if result['label'] == 'LABEL_0':
        return f'The text transcribed non-hateful with a {score:.01f}% certainty'
    else:
        return f'The text transcribed is hateful with a {score:.01f}% certainty'

def predict_audio_vs(audio):
    voices = split_voices(audio)
    text1=pipe_audio(voices[0][0], generate_kwargs={ "task": "transcribe", "language": "<|ar|>" })['text']
    text2=pipe_audio(voices[0][1], generate_kwargs={ "task": "transcribe", "language": "<|ar|>" })['text']
    result1=pipe_HSC(text1)[0]
    label1 = describe_result(result1)
    result2=pipe_HSC(text2)[0]
    label2 = describe_result(result2)
    return (16_000, voices[0][0]), text1, label1, (16000, voices[0][1]), text2, label2

```

```python
with gr.Blocks() as demo:
    with gr.Tab("Text Classification"):
        text_input = gr.Textbox(label="Input Text")
        text_output = gr.Textbox(label="Output Classification")
        text_button = gr.Button("Classify")
    with gr.Tab("Audio Classification"):
        audio_input=gr.Audio(label='Audio Input', source='microphone', type='filepath')
        audio_transcription_output=gr.Textbox(label="Audio Transcription")
        audio_classification_output=gr.Textbox(label="Output Classification")
        audio_button = gr.Button("Classify")
    with gr.Tab("Multi-Speaker Classification"):
        audio_mix_input=gr.Audio(label='Audio Input', source='microphone', type='filepath')
        with gr.Row():
            with gr.Column():
                speech1=gr.Audio(label='Speech 1', interactive=False)
                audio_transcription1_output=gr.Textbox(label="Transcription 1")
                label1=gr.Textbox(label="Classification 1")
            with gr.Column():
                speech2=gr.Audio(label='Speech 2', interactive=False)
                audio_transcription2_output=gr.Textbox(label="Transcription 2")
                label2=gr.Textbox(label="Classification 1")
        vs_audio_button = gr.Button("Classify")

        

    text_button.click(predict_text, 
                      inputs=text_input, 
                      outputs=text_output
                     )
    audio_button.click(predict_audio, 
                       inputs=audio_input, 
                       outputs=[audio_transcription_output,audio_classification_output]
                      )
    vs_audio_button.click(predict_audio_vs, 
                       inputs=audio_mix_input, 
                       outputs=[speech1, audio_transcription1_output, label1, speech2, audio_transcription2_output,label2]
                      )

demo.launch()
```
